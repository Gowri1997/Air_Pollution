```python
pollution_metro['AQI Value']
```

```python
import numpy as np
X=['Delhi','Hyderabad','Pune','Bangalore','Surat']

Y_AQI=pollution_metro['AQI Value']
Y_CO=pollution_metro['CO AQI Value']
Y_O3=pollution_metro['Ozone AQI Value']
Y_NO2=pollution_metro['NO2 AQI Value']
Y_PM=pollution_metro['PM2.5 AQI Value']

X_axis=np.arange(len(X))

plt.bar(X_axis-0.2,Y_AQI,0.1,label='AQI value')
plt.bar(X_axis-0.1,Y_CO,0.1,label='CO AQI')
plt.bar(X_axis-0,Y_O3,0.1,label='Ozone AQI')
plt.bar(X_axis+0.1,Y_NO2,0.1,label='NO2 AQI')
plt.bar(X_axis+0.2,Y_PM,0.1,label='PM2.5 AQI')  

plt.xticks(X_axis,X)
plt.ylabel('AQI Value')
plt.legend()
plt.show()
```
